```python
import skimage
import numpy as np
import copy
import matplotlib.pyplot as plt
from tqdm import tqdm
from skimage import io
from skimage.draw import disk
```

```python
def get_data_from_file(path) :
    file_to_read = open(path, "r")
    ret_data = np.loadtxt(file_to_read)
    file_to_read.close()
    return ret_data
```

```python
def draw_plot(title, data_3, data_8, data_15) :
    plt.figure()
    plt.title(title)
    plt.xlabel("liczba kroków")
    plt.ylabel("średni fitness (logarytm dziesiętny)")
    plt.plot(np.log10(data_3), label = "K = 3")
    plt.plot(np.log10(data_8), label = "K = 8")
    plt.plot(np.log10(data_15), label = "K = 15")
    plt.legend()
    plt.show()
```

```python
path_3_kot = "color-tiles/3col/col3kot.txt"
path_8_kot = "color-tiles/8col/col8kot.txt"
path_15_kot = "color-tiles/15col/col15kot.txt"

path_3_plaza = "color-tiles/3col/col3plaza.txt"
path_8_plaza = "color-tiles/8col/col8plaza.txt"
path_15_plaza = "color-tiles/15col/col15plaza.txt"

path_3_kwiat = "color-tiles/3col/col3kwiat.txt"
path_8_kwiat = "color-tiles/8col/col8kwiat.txt"
path_15_kwiat = "color-tiles/15col/col15kwiat.txt"
```

```python
data_3_kot = get_data_from_file(path_3_kot)
data_8_kot = get_data_from_file(path_8_kot)
data_15_kot = get_data_from_file(path_15_kot)

draw_plot("Kot średni, b = 14", data_3_kot, data_8_kot, data_15_kot)
```

```python
data_3_plaza = get_data_from_file(path_3_plaza)
data_8_plaza = get_data_from_file(path_8_plaza)
data_15_plaza = get_data_from_file(path_15_plaza)

draw_plot("Plaża średnia, b = 7", data_3_plaza, data_8_plaza, data_15_plaza)
```

```python
data_3_kwiat = get_data_from_file(path_3_kwiat)
data_8_kwiat = get_data_from_file(path_8_kwiat)
data_15_kwiat = get_data_from_file(path_15_kwiat)

draw_plot("Kwiat średni, b = 28", data_3_kwiat, data_8_kwiat, data_15_kwiat)
```
